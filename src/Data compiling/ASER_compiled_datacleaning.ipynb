```python
import pandas as pd
```

```python
# 1. Load your Stata file
# Replace this with your actual path
file_path = "/Users/vidhi/college baby/sem 5/Dev/End term paper/Do/ASER_Panel_districts_cleanfs.dta"
print("Loading data... this might take a moment for 8M rows.")
df = pd.read_stata(file_path)
```

```python
# --- STEP 1: Basic String Hygiene ---
print("Step 1: Removing quotes, spaces, and fixing capitalization...")

# Convert to string, remove quotes, strip whitespace, title case
df['district_name'] = df['district_name'].astype(str)
df['district_name'] = df['district_name'].str.replace('"', '', regex=False)
df['district_name'] = df['district_name'].str.strip().str.title()

# Remove periods (e.g., "S.A.S. Nagar" -> "Sas Nagar") for consistency
df['district_name'] = df['district_name'].str.replace('.', '', regex=False)
```

```python
# --- STEP 1: Enhanced String Cleaning (Fixes Chhattisgarh & Punjab) ---
print("Step 1: Cleaning punctuation and spacing...")

df['district_name'] = df['district_name'].astype(str)
# Remove quotes
df['district_name'] = df['district_name'].str.replace('"', '', regex=False)

# Replace Underscores and Dashes with a SPACE
df['district_name'] = df['district_name'].str.replace('_', ' ', regex=False)
df['district_name'] = df['district_name'].str.replace('-', ' ', regex=False)

# Clean up extra spaces (e.g. "Janjgir  Champa" -> "Janjgir Champa")
df['district_name'] = df['district_name'].str.replace(r'\s+', ' ', regex=True)
df['district_name'] = df['district_name'].str.strip().str.title()
df['district_name'] = df['district_name'].str.replace('.', '', regex=False)
```

```python
# --- STEP 2: Find and Print Conflicts ---
print("\nChecking for inconsistencies...\n")

# Group by State and District Code, then count unique spellings of the name
grouped = df.groupby(['state_name', 'district_code'])['district_name'].unique()

# Filter to keep only codes that have MORE THAN ONE name associated with them
conflicts = grouped[grouped.apply(len) > 1]

if len(conflicts) == 0:
    print("Great news! No conflicts found. Every district code maps to exactly one name.")
else:
    print(f"Found {len(conflicts)} district codes with multiple name spellings:\n")
    print(f"{'State':<20} | {'Code':<6} | {'Variations Found'}")
    print("-" * 60)
    
    for (state, code), names in conflicts.items():
        # Join the different names into a single string to display
        names_str = ", ".join(names)
        print(f"{state:<20} | {code:<6} | {names_str}")

    print("-" * 60)
    print("\nTo fix these automatically, run the next block of code (the 'Smart Trick').")
```
